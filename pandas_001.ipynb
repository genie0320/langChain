#### 2. Index
데이터에 접근할 수 있는 주소 값.

```python
# 원래 col no.가 붙지만, 위와 같이 index값을 별도로 지정해줬다면, 해당 index값만 모아서 불러올 수도 있다.
'''
df.index
'''

# index 이름설정 > index 열 명명하기
'''
df.index.name = 'index_name'
'''
# index 초기화
'''
df.reset_index() > 새로 디폴트 번호 col이 제일 앞에 '추가' 생성됨.
df.reset_index(drop = True) > 현재의 index열을 디폴트값으로 '교체'.
df.reset_index(drop = True, inplace=True) > 변경된 데이터를 저장함.
'''
# index 설정 > 데이터의 한 열을 index로 지정
'''
df.set_index('key')
df.set_index('key', inplace=True) > 변경한 대로 데이터를 저장
'''
# index 정렬
'''
df.sort_index() > 오름차순 정렬
df.sort_index(ascending=False) > 내림차순 정렬
'''
```

```python
df['키'].describe()

df['키'].min() # 168
df['키'].nlargest(3) # 큰 순서로 3개
df['키'].mean() # 평균
df['키'].sum()
df['SW특기'].count() # NaN 제외 숫자
df['학교'].unique() # array(['북산고', '능남고'], dtype=object)
df['학교'].nunique() # 고유값의 갯수를 가져옴.
```

#### DataFrame 확인

1) 데이터의 전체 모양 : df.info()
데이터 갯수, Null 여부 등 확인

2) 데이터의 계산값 : df.describe()
숫자적으로 계산될 수 있는 데이터의 최소/최대, 편차치 등을 보여준다.

- count: 빈 값을 제외한 데이터의 개수
- mean: 데이터의 평균
- std: 데이터의 표준 편차
- min: 데이터의 최소값
- 25%: 데이터의 1사분위수 (25% 값)
- 50%: 데이터의 중앙값 (50% 값)
- 75%: 데이터의 3사분위수 (75% 값)
- max: 데이터의 최대값   

3) 데이터 샘플값 : df.head() > 최초 5개의 값을 가져온다.
- df.head(n) > n 개 데이터 가져오기
- df.tail() > 마지막 몇개 *가져오기*
